# Numpy Dersleri

Numpy'da tum elemanlar ayni tiptedir.
Bu sayede daha az yer kaplar ve cok daha hizli calisir.

```python
import numpy as np
```

#### Liste veya Tuple'dan Array Oluşturma

```python
# create a np array

arr1 = np.array([1, 3, 87, 44])

type(arr1)
```

```python
# python array type

list1 = [5, 7, 12]

type(list1)
```

```python
# python array to np array 

arr2 = np.array(list1)

type(arr2)
```

```python
# tuple to np array

arr3 = np.array((10, 20, 30))

type(arr3)
```

#### Çok Boyutlu Array Nasıl Oluşturulur?

```python
# 2x3 array

arr4 = np.array([[1, 2, 3], [4, 5, 6]])
type(arr4)
```

```python
# Shape, array boyutu

arr4.shape
```

#### Sıfır, Bir Değerleri ile Dolu ve Boş Array Oluşturma 

```python
z = np.zeros((2, 3)) # 2x3
z
```

```python
o = np.ones((3, 2)) # 3x2
o
```

```python
e = np.empty((2, 2)) # 2x2 boş (rastgele değer gelebilir)
e
```

```python
e[0][0] = 40
e
```

```python
w = np.zeros(5) # Varsayılan olarak dtype = float olur yani 0.0 
w
```

```python
w2 = np.zeros(5, dtype=int)
w2
```

#### Sayı Aralıkları ile Array Oluşturma

```python
a1 = np.arange(0, 10, 2) # 0'dan 10'a kadar 2'şer 2'şer git.
a1
```

```python
s1 = np.linspace(0, 1, 5) # 0 ile 1 arasında 5 sayı oluştur.
s1
```

#### Rastgele (Random) Array Oluşturma

```python
r1 = np.random.rand(2, 3) # 2x3, 0-1 arasında float array
r1
```

```python
r2 = np.random.randint(0, 10, (2, 3)) # 2x3, 0-9 arası integer.
r2
```

#### dtype (Veri Tipi) ile Array Oluşturma

```python
arr6 = np.array([1, 2, 3], dtype=float)
print(arr6)
```

```python
arr_bool = np.array([1, 0, 1], dtype=bool)
arr_bool
```
